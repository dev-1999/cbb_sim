```python
from models import generate_historical_df
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, roc_curve
```

```python
df = generate_historical_df()
```

```python
df['combo_year'] = df.Label_Lower.apply(lambda x: str(x)[-4:])
```

```python
df['made_tourney'] = df.S.apply(lambda x: 1 if x < 17 else 0)
```

```python
scaler = StandardScaler()
```

```python
df = df.loc[df.Label_Lower != 'louisville2016']
```

```python
h = df[['mean_pythag','mean_adjOE','mean_adjDE','mean_tempo']]
```

```python
scaler.fit(h)
h = scaler.transform(h)
```

```python
label_binary = df['made_tourney']
```

```python
labels = df['S']
```

```python
x_train, x_test, y_train, y_test = train_test_split(h, labels, test_size=0.25, random_state = 1)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

r2_scores = []
```

```python
def gen_predictions(a, b, n):
    neigh = NearestNeighbors(n_neighbors = n)
    neigh.fit(x_train)
    y_pred = []
    for i in x_test:
        #print(i)
        nb = neigh.kneighbors([i])
        #print(nb)
        distances = nb[0][0].tolist()
        inv_dist = [(a/d**b) for d in distances]
        matches = nb[1][0].tolist()
        match_labels = [y_train[m] for m in matches]
        #print(match_labels)
        pred = 0
        for i in range(len(match_labels)):
            pred += inv_dist[i] * match_labels[i]
        pred = pred/sum(inv_dist)
        y_pred.append(pred)
    return y_pred
```

```python
r2_scores = []
for neighbor_count in range(35,75,5):
    predictions = gen_predictions(1,1,neighbor_count)
    score = (r2_score(y_test,predictions))
    r2_scores.append(score)
    #print(score)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=range(35,75,5),y=r2_scores)
plt.show()
```

```python
r2_scores = []
for b_parameter in range(5,25):
    predictions = gen_predictions(1,b_parameter/10,45)
    score = (r2_score(y_test,predictions))
    r2_scores.append([score, b_parameter/10, neighbor_count])
```

```python
plt.scatter(x=[i/10 for i in range(5,25)], y=[i[0] for i in r2_scores])
```

```python
import io
import base64
def list_to_hist(lst):
    fig, ax = plt.subplots()
    plt.style.use('default')
    data = []
    img = io.BytesIO()
    for i in range(len(lst)):
        for j in range(int(lst[i]*1000)):
            data.append(i+1)
    ax.hist(data, bins=np.arange(0, len(lst)+1)+0.5, density = True, color="#364e8f")
    ticklabels = [str(i) for i in range(1,18)]
    ticklabels[-1] = 'Miss'
    plt.xticks(range(1,18), ticklabels, size='small')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.savefig(img, format='png', bbox_inches='tight', transparent="True")
    return base64.b64encode(img.getvalue()).decode()
```
